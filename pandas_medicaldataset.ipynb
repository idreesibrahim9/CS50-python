```python
import matplotlib.pyplot as plt
import seaborn as seb
plt.scatter(df['age'], df['charges'])
x=df['age']
y=df['charges']
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot of Age vs Charges')
plt.show()
```

```python
plt.scatter(df['bmi'], df['charges'])
x=df["bmi"]
y=df["charges"]
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("Scatter Plot of BMI vs Charges")
plt.show()
```

```python
df['age'].value_counts().plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.stem(df["age"], df["charges"])
x=df["age"]
y=df["charges"]
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Stem Plot of Age vs Charges")
plt.show()
```

```python
plt.hist(df["age"])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.hist2d(df["age"], df["charges"])
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("2D Histogram of Age vs Charges")
plt.show()
```

```python
plt.hexbin(df["age"], df["charges"], gridsize=20, cmap='Blues')
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Hexbin Plot of Age vs Charges")
plt.show()
```

```python
seb.scatterplot(x=df["age"], y=df["charges"])
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Scatter Plot of Age vs Charges")
plt.show()
```

```python
seb.regplot(x=df["age"], y=df["charges"])
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Regression Plot of Age vs Charges")
plt.show()
```

```python
seb.boxplot(x=df["smoker"], y=df["charges"])
plt.xlabel("Smoker (0=No, 1=Yes)")
plt.ylabel("Charges")
plt.title("Box Plot of Charges by Smoker Status")
plt.show()
```

```python
seb.violinplot(x=df["smoker"], y=df["charges"])
plt.xlabel("Smoker (0=No, 1=Yes)")
plt.ylabel("Charges")
plt.title("Violin Plot of Charges by Smoker Status")
plt.show()
```

```python
seb.histplot(df["age"], kde=True)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Histogram of Age")
plt.show()
```

```python
seb.kdeplot(df["age"], fill=True)
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("KDE Plot of Age")
plt.show()
```
